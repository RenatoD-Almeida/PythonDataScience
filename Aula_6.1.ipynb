# Analise de dados de Filmes exibidos no Brasil de 2009 a 2022
Com base nos dados da Ancine (http://oca.ancine.gov.br)
• Vamos analisar os filmes exibidos no brasil
arquivo -> 2121.csv

```python
import pandas as pd
%matplotlib inline

df = pd.read_csv("2121.csv", sep=';', encoding='cp1252', usecols = [0, 1, 3, 4, 6], skipfooter = 27)
# skiprows comando para pular linhas. skipfooter = tira as ultimas linhas!
```

```python
df.tail(27)
```

```python
df_porAno = df.groupby('Ano de Lançamento')['Gênero'].value_counts()
df_porAno
```

```python
# Quantos filmes de Ficção foram lançados por ano: Linha
df[df['Gênero'] == "Ficção"]['Ano de Lançamento'].value_counts()
```

```python
# Quantos filmes de Ficção foram lançados por ano: Tabela
gr_ficcao = df[df['Gênero'] == "Ficção"]['Ano de Lançamento'].value_counts()
gr_ficcao.plot.bar()
```
