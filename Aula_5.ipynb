```python
import pandas as pd
%matplotlib inline

df = pd.read_csv('cursos-prouni.csv')
```

```python
df.head()
```

```python
df_por_mensalidade = df[['nome' ,'mensalidade']]
grafico_por_mensalidade = df_por_mensalidade[df['mensalidade'] > 5000]['nome'].value_counts()
```

```python
# gráfico sobre a quantidade de cursos acima de R$5000,00
grafico_por_mensalidade.plot.barh()
```

```python
#Uf com quantidade de cursos acima de 1000

cursos_por_uf = df['uf_busca'].value_counts()
cursos_por_uf[cursos_por_uf > 1000].plot.pie()
```

```python
df.dtypes
```

```python
# Mensalidade mais cara por UF
df_uf_mensalidade = df.groupby('uf_busca')['mensalidade'].max()
df_uf_mensalidade
```

```python
# Mensalidade mais barata por UF
df_uf_mensalidade = df.groupby('uf_busca')['mensalidade'].min()
df_uf_mensalidade
```

```python
# Mostrar todos os dados dos cursos mais caros de cada uf
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
# Mensalidades entre R$500,00 e R$1000,00

df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)][['nome', 'mensalidade']]
```
